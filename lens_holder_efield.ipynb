```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_file = r"C:\Users\yuhan\work\microspheres\data\lens_hodler_efield_sim.csv"
data = np.genfromtxt(data_file, dtype=np.float64, skip_header=8, delimiter=',').T
```

```python
plt.plot(data[0], data[1])
plt.xlabel('Distance (mm)')
plt.ylabel('Electric potential (V)')
```

```python
plt.plot(data[0], 1000*np.gradient(data[1], data[0]))
plt.xlabel('Distance (mm)')
plt.ylabel('Electric Field (mV/mm)')
```
