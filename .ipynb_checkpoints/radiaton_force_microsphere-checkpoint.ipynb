## Refractive index

Plot the refracetive index of fused silica at different wavelengths.

```python
# Import the refractive index data for fused silica
sio2 = np.genfromtxt('https://refractiveindex.info/tmp/data/main/SiO2/Malitson.txt', delimiter='\t')

# Data is stacked so need to rearrange
# SiO2 does not have complex refractive index in the database
N = len(sio2)
sio2_lam = sio2[1:N, 0]  # wavelength in microns
sio2_mre = sio2[1:N, 1]

plt.figure(figsize=(6, 4))
plt.scatter(sio2_lam*1000,sio2_mre,s=2,color='blue')
plt.xlim(300, 1000)
plt.ylim(0.5,2)

plt.xlabel('Wavelength (nm)')
plt.ylabel('Refractive Index')
plt.text(350, 1.67, '$m_{re}$', color='blue', fontsize=14)

plt.title('Refractive Index of Fused Silica')
plt.show()
```

## Radiation force

`miepython.mie` calculate the Mie efficiencies given the complex index of refraction ($m$) and the dimensionless size parameters of the sphere ($x = 2 \pi r / \lambda$ where $r$ is radius and $\lambda$ is the wavelength of the incident light).

The formulas of the Mie efficiencies can be found in many texts e.g. Kerker or Bohren & Hauffman, and we will record them here for completeness. We will use the extinction coefficient
$$
    Q_{ext} = \frac{2}{x^2} \sum^{N}_{n=1} (2n+1) \text{Re}(a_n + b_n),
$$

the scattering coefficient
$$
    Q_{sca} = \frac{2}{x^2} \sum^{N}_{n=1} (2n+1) (|a_n|^2 + |b_n|^2),
$$

and the scattering anisotropy (averaged cosine)
$$
    g = \frac{4}{x^2 Q_{sca}} \sum^{N}_{n=1} [ \frac{n(n+2)}{(n+1)} \text{Re}(a_n a_{n+1}^{*} + b_n b_{n+1}^*) + \frac{2n+1}{n(n+1)} \text{Re}(a_n b_n^*)].
$$
In the expressions above, $a_n$ and $b_n$'s are the Mie coefficients labeling the partial wave modes.

The radiation pressure efficiency is given by
$$
    Q_{rad} = Q_{ext} - g Q_{sca},
$$
which is the fraction of momentum given to the sphere in the incident direction.

The force on the sphere is
$$
    F = \frac{p}{c} Q_{rad},
$$
where $p$ is the power incident on the area of the sphere.

Now we calculate the radiation force on microspheres with different size parameters. Compare the result to Fig.3 of Barker 2010.

```python
# Size parameters of interest
# Need high enough resolution to resolve the peaks
x = np.linspace(39, 41, 400000)
m = 1.45

qext, qsca, qback, g = miepython.mie(m,x)
qpr  = qext - g*qsca

p = 0.01  # incident power in watts on the cross-sectional area
c = 3e8   # speed of light (m/s)
f = qpr * p/c
```

```python
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111)

ax.plot(x, f*1e12, color='blue')
ax.set_xlim(39, 41)
ax.set_ylim(5, 21)
ax.set_xlabel('$x=2 \pi r \ / \ \lambda$; Dimensionless size param', fontsize=14)
ax.set_ylabel('Radiation Force (pN)', fontsize=14)
ax.set_title('Force on a fused silica microsphere (m=1.45, p=10 mW)', fontsize=14)
plt.show()
```

Calculate the force on spheres of different sizes. Again, we assume 10 mW power incident on the cross section of the spheres. Compare this to Sarah's results.

```python
def get_qpr(r, lam, m):
    x = 2 * np.pi * r / lam
    qext, qsca, qback, g = miepython.mie(m,x)
    qpr  = qext - g*qsca
    return qpr
```

```python
import matplotlib as mlp
from cycler import cycler
mlp.rcParams['axes.prop_cycle'] = cycler(color='bgrcmyk')

fig, axes = plt.subplots(1, 2, figsize=(20, 8))

lam = np.linspace(0.778, 0.782, 40000)

# Assume constant indx of refraction m = 1.45
m0 = np.full_like(lam, 1.45)
# Or interpolation
m1 = np.interp(lam, sio2_lam, sio2_mre)

p = 0.01 # watts on the spheres
c = 3e8

radius = np.array([5, 10, 15, 20, 25]) / 2

for i, m in enumerate([m0, m1]):
    for r in radius:
        qpr = get_qpr(r, lam, m)
        f = qpr * p/c
        axes[i].plot(lam * 1000, f*1e12, label=f'{r*2:.0f} $\mu$m')
        
axes[0].legend(fontsize=14, frameon=False)
axes[0].set_xlabel('Wavelength (nm)', fontsize=16)
axes[0].set_ylabel('Radiation Force (pN)', fontsize=16)
axes[0].set_title('m=1.45, p=10 mW', fontsize=16)
axes[1].set_title('m by interpolation, p=10 mW', fontsize=16)
for ax in axes:
    ax.set_xlim(778, 782)
    ax.tick_params(axis='x', labelsize=14)
    ax.tick_params(axis='y', labelsize=14)
plt.show()
```

Radiation force as a function of particle radius (780 nm, 10 mW power incident on the sphere)

```python
p = 0.01
c = 3e8
radius = np.linspace(4.7, 5.2, 500)
lam = np.full_like(radius, 0.780)
m0 = np.full_like(radius, 1.45)

qpr = get_qpr(radius, lam, m0)
```
