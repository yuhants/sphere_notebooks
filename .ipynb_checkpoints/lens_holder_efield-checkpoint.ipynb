```python
import numpy as np
import matplotlib.pyplot as plt
```

## Long version

```python
data_file = r"C:\Users\yuhan\work\microspheres\data\lens_hodler_efield_sim.csv"
data = np.genfromtxt(data_file, dtype=np.float64, skip_header=8, delimiter=',').T
```

```python
fig, ax = plt.subplots(2, 1, figsize=(10, 7))

ax[0].plot(data[0], data[1])
ax[1].plot(data[0], 1000*np.gradient(data[1], data[0]))

ax[0].set_ylabel('Potential (V)', fontsize=14)
ax[1].set_xlabel('Distance (mm)', fontsize=14)
ax[1].set_ylabel('E Field (mV/mm)', fontsize=14)
```

## Short version

```python
data_file = r"C:\Users\yuhan\work\microspheres\data\lens_hodler_efield_sim_shortver.csv"
data = np.genfromtxt(data_file, dtype=np.float64, skip_header=8, delimiter=',').T
```

```python
fig, ax = plt.subplots(2, 1, figsize=(8, 4))

ax[0].plot(data[0], -1*data[1])
ax[1].plot(data[0], -1000*np.gradient(data[1], data[0]))

ax[0].set_ylabel('Electric potential (V)')
ax[1].set_xlabel('Distance (mm)')
ax[1].set_ylabel('Electric Field (mV/mm)')
```
