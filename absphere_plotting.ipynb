```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import matplotlib.cm as cm
from matplotlib.colors import LogNorm
```

```python
def read_field(file):
    """Get E field data and (x_min, x_max, y_min, y_max)"""
    with open(file) as f:
        lines = f.readlines()
        
    size = len(lines) - 15
    
    data = np.empty(shape=(size, 3), dtype=np.float64)
    for i, line in enumerate(lines[2:]):
        data[i] = line.split()
    
    data = data.T
    ext = (data[0].min(), data[0].max(), data[1].min(), data[1].max())
    d = data[2].reshape((xsize, xsize))
    
    return d, ext
```

```python
def read_force(file):
    """Get particle position/size and Fx, Fy, Fz as a numpy array"""
    with open(file) as f:
        lines = f.readlines()
        
    size = len(lines) - 2
    data = np.empty(shape=(size, 4), dtype=np.float64)
    for i, line in enumerate(lines[2:]):
        data[i] = line.split()
    data = data.T
    
    return data
```

```python
def get_peak(r, fz):
    vmax = np.amax(fz)
    imax = np.argmax(fz)
    return [r[imax], vmax]
```
