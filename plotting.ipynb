```python
import numpy as np
import matplotlib.pyplot as plt
```

Plot a Lorentizian
$$
f(x) = \frac{A \gamma^2}{\gamma^2 + (x - x_0)^2}
$$

```python
A = 10000
gamma = 0.5
x0 = 5

x = np.linspace(0, 10, 1000)
y = A*gamma*gamma / (gamma*gamma + np.square(x-x0))
```

```python
fig = plt.figure(figsize = (7, 6))
ax = fig.add_subplot()

ax.plot(x, y, 'k')
ax.set_yticklabels([])
ax.set_xticklabels([])

ax.vlines(4.5, 0, 10000, colors='b')
ax.vlines(5, 0, 10000, colors='k', linestyle='--')
ax.text(0.05, 7000, 'Laser Frequency', color='b', fontsize=18)
ax.text(6.5, 2000, 'Resonant \nFrequency', color='k', fontsize=18)

ax.set_xlabel('Frequency', fontsize=18)
ax.set_ylabel('Radiation Force', fontsize=18)
```
